```python
# Detection outliers
def detect_outliers_iqr(data, threshold=1.5):
    
    q25, q75 = np.percentile(data, 25), np.percentile(data, 75)
    iqr = q75 - q25
    lower_bound = q25 - (threshold * iqr)
    upper_bound = q75 + (threshold * iqr)
    outliers = data[(data < lower_bound) | (data > upper_bound)]
    return outliers
```

```python
columns=car_insurance.describe().columns
columns
```

```python
# Detecting outliers and printing them 
column=[]
no_outliers=[]
outliers_percent=[]
no_unique_outlier=[]
actual_unique_no=[]
data_min=[]
data_max=[]
type_outliers=[]

for col in columns:
    outliers = detect_outliers_iqr(car_insurance[col])
    column.append(col)
    no_outliers.append(len(outliers))
    percentage=round((len(outliers)/len(car_insurance))*100,2)
    outliers_percent.append(percentage)
    no_unique_outlier.append(outliers.nunique())
    actual_unique_no.append(car_insurance[col].nunique())
    data_min.append(car_insurance[col].min())
    data_max.append(car_insurance[col].max())
    type_outliers.append(outliers.unique())

outliers_dic={"column":column,
              "no_outliers":no_outliers,
              "no_unique_outlier":no_unique_outlier,
              "outliers_percent":outliers_percent,
              "actual_unique_no":actual_unique_no,
              "data_min":data_min,
              "data_max":data_max,
              "type_outliers":type_outliers
             }

outliers_df=pd.DataFrame(outliers_dic)

display(outliers_df.style.set_properties(**{'text-align': 'left'}))
    
```

```python
# Creating line plot 
def line_plot(data, feature, ax):
    
    # preparing data for line plot
    claim_counts = data.groupby(feature)["is_claim"].value_counts().unstack().fillna(0)
    
    # Plotting line plot (value count)
    claim_counts.plot(kind="line", marker='o', ax=ax) 
    ax.set_xlabel(f"{feature}")
    ax.set_ylabel("Count")
    ax.set_title(f"Claims vs. Non-claims by {feature}")
    ax.legend(["Non-claim", "Claim"], loc="upper left")
    ax.set_xticks(range(len(claim_counts.index)))
    ax.set_xticklabels(claim_counts.index, rotation=45)
    ax.grid(True)
    
    # Getting maximim and minimum calues (count)
    for claim_type in claim_counts.columns:
        max_idx_label = claim_counts[claim_type].idxmax()
        min_idx_label = claim_counts[claim_type].idxmin()
        max_val = claim_counts[claim_type].max()
        min_val = claim_counts[claim_type].min()
        max_idx = claim_counts.index.get_loc(max_idx_label)
        min_idx = claim_counts.index.get_loc(min_idx_label)
        
        # Plotting maximim value 
        ax.plot(max_idx, max_val, 'go')
        ax.annotate(f'Max: {max_val}', xy=(max_idx, max_val), xytext=(max_idx, max_val + 1),
                    arrowprops=dict(facecolor='black', arrowstyle="->"), ha='center')
        
        # Plotting minimum value
        ax.plot(min_idx, min_val, 'ro')
        ax.annotate(f'Min: {min_val}', xy=(min_idx, min_val), xytext=(min_idx, min_val - 1),
                    arrowprops=dict(facecolor='black', arrowstyle="->"), ha='center')
    

# Creating line plots of relative percentage
def line_plot_percentage(data, feature, ax):
    
    # Preparing data for plottig
    claim_counts = data.groupby(feature)["is_claim"].value_counts(normalize=True).unstack().fillna(0)
    claim_counts = round(claim_counts * 100, 3)
    
    # plotting relative percentage data 
    claim_counts.plot(kind="line", marker='o', ax=ax) 
    ax.set_xlabel(f"{feature}")
    ax.set_ylabel("Percentage")
    ax.set_title(f"Claims vs. Non-claims by {feature} (percentage)")
    ax.legend(["Non-claim", "Claim"], loc="upper left")
    ax.set_xticks(range(len(claim_counts.index)))
    ax.set_xticklabels(claim_counts.index, rotation=45)
    ax.grid(True)
    
    # Removing legend
    ax.get_legend().remove()
    
    # Getting maximum and minimum relative percentage
    for claim_type in claim_counts.columns:
        max_idx_label = claim_counts[claim_type].idxmax()
        min_idx_label = claim_counts[claim_type].idxmin()
        max_val = claim_counts[claim_type].max()
        min_val = claim_counts[claim_type].min()
        max_idx = claim_counts.index.get_loc(max_idx_label)
        min_idx = claim_counts.index.get_loc(min_idx_label)
        
        # Plotting maximum relative percentage
        ax.plot(max_idx, max_val, 'go')
        ax.annotate(f'Max: {max_val}%', xy=(max_idx, max_val), xytext=(max_idx, max_val + 1),
                    arrowprops=dict(facecolor='black', arrowstyle="->"), ha='center')
        
        # Plotting minimum relative percenatge
        ax.plot(min_idx, min_val, 'ro')
        ax.annotate(f'Min: {min_val}%', xy=(min_idx, min_val), xytext=(min_idx, min_val - 1),
                    arrowprops=dict(facecolor='black', arrowstyle="->"), ha='center')
```

```python
# Creating stacked bar plot
def stacked_bar(data, feature, ax):
    # Preparing data for stacked bar plot
    claim_counts = data.groupby(feature)["is_claim"].value_counts().unstack().fillna(0)
    
    # Plotting stacked bar plot (value count)
    claim_counts.plot(kind="bar", stacked=True, ax=ax)
    ax.set_xlabel(f"{feature}")
    ax.set_ylabel("Count")
    ax.set_title(f"{feature}")
    ax.legend(["Non-claim", "Claim"], loc="upper left")
    ax.set_xticks(range(len(claim_counts.index)))
    ax.set_xticklabels(claim_counts.index, rotation=45)
    ax.grid(True)
    
    # Annotating bar chart
    for p in ax.patches:
        height = p.get_height()
        if height > 0: 
            ax.annotate(f'{height:.2f}', 
                        (p.get_x() + p.get_width() / 2., p.get_y() + height / 2.), 
                        ha='center', va='center', xytext=(0, 0), textcoords='offset points')
    
# Creating stacked bar plots of relative percentage
def stacked_bar_percentage(data, feature, ax):
    # Preparing data for plotting
    claim_counts = data.groupby(feature)["is_claim"].value_counts(normalize=True).unstack().fillna(0)
    claim_counts = round(claim_counts * 100, 3)
    
    # Plotting relative percentage data
    claim_counts.plot(kind="bar", stacked=True, ax=ax)
    ax.set_xlabel(f"{feature}")
    ax.set_ylabel("Percentage")
    ax.set_title(f"{feature} (percentage)")
    #ax.legend(["Non-claim", "Claim"], loc="upper left")
    ax.set_xticks(range(len(claim_counts.index)))
    ax.set_xticklabels(claim_counts.index, rotation=45)
    ax.grid(True)
    
    # Removing legend
    ax.get_legend().remove()
    
    # Annotating bar chart
    for p in ax.patches:
        height = p.get_height()
        if height > 0: 
            ax.annotate(f'{height:.2f}%', 
                        (p.get_x() + p.get_width() / 2., p.get_y() + height / 2.), 
                        ha='center', va='center', xytext=(0, 0), textcoords='offset points')
    
```

```python
def compare_two_features(data, feature1, feature2):
    result = data.groupby([feature1, feature2]).size().unstack(fill_value=0)
    return result
```

```python
# Dividing x and y data sets
x = car_insurance.drop('is_claim', axis=1)
y = car_insurance['is_claim']
```

```python
# Defining columns for scaling and encoding
numerical_feature = x.select_dtypes(include=['int64','float64']).columns
categorical_feature = x.select_dtypes(include=['object']).columns
categorical_feature_ohe = x.select_dtypes(include=['object']).columns[x[categorical_feature].nunique() > 2]
categorical_feature_label=[col for col in categorical_feature if col not in categorical_feature_ohe ]

print("Numerical Features")
print(numerical_feature)
print("-"*100)
print("Categorical Features")
print(categorical_feature)
print("-"*100)
print("Categorical Features for one label encode")
print(categorical_feature_label)
print("-"*100)
print("Categorical Features for one hot encoder")
print(categorical_feature_ohe)
```

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder, OneHotEncoder

def scaling_encoding(data,scalrf,labelf,ohef):
    scaler = StandardScaler()
    ohe = OneHotEncoder()

    # Perform Standard Scaling on numerical features
    for col in scalrf:
        data[col] = scaler.fit_transform(data[[col]])

    # Initialize LabelEncoder 
    label_encoder = LabelEncoder()
    for col in labelf:
            data[col] = label_encoder.fit_transform(data[col])

    # Perform One-Hot Encoding 
    encoded_features = ohe.fit_transform(data[ohef])
    encoded_df = pd.DataFrame(encoded_features.toarray(), columns=ohe.get_feature_names_out(ohef))

    encoded = pd.concat([data.drop(columns=ohef), encoded_df], axis=1)

    # Display the encoded DataFrame
    return encoded
```

```python
encoded_car_insuarance=scaling_encoding(car_insurance,numerical_feature,categorical_feature_label,categorical_feature_ohe)
encoded_car_insuarance
```

```python
from scipy.stats import pearsonr

correlations = {}

# Iterate over each column in the DataFrame
for col in encoded_car_insuarance.columns:
    corr, p_value = pearsonr(encoded_car_insuarance[col], encoded_car_insuarance['is_claim'])
    correlations[col] = {'Correlation': corr, 'P-value': p_value}

correlations_df = pd.DataFrame.from_dict(correlations, orient='index')
correlations_df['P-value'] = correlations_df['P-value'].map(lambda p: '{:.10f}'.format(p))

correlations_df
```

```python
# Splitting data for model training
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(X_resampled, y_resampled,test_size=0.2,random_state=42)
print(x_train.shape)
print(x_test.shape)
print(y_train.value_counts())
print(y_test.value_counts())
```

```python
# Creating list for better representation of model results
model_name_l=[]
train_accuracy_l=[]
test_accuracy_l=[]
train_f1_score_l=[]
test_f1_score_l=[]
train_precision_l=[]
test_precision_l=[]
train_recall_l=[]
test_recall_l=[]
train_auc_roc_l=[]
test_auc_roc_l=[]
```

```python
from sklearn.metrics import (confusion_matrix, ConfusionMatrixDisplay, accuracy_score, 
                             precision_score, recall_score, f1_score, roc_auc_score, roc_curve, auc, classification_report)
import matplotlib.pyplot as plt

# creating a function to train, validate, test, and print roc_auc curve
def model_training(ml_model, x_train, x_test, y_train, y_test):
    
    # Training the model
    model = ml_model
    model.fit(x_train, y_train)
    
    # Predicting with the model
    y_train_predict = model.predict(x_train)
    y_test_predict = model.predict(x_test)
    
    # Validating the model with training performance metrics
    train_accuracy = accuracy_score(y_train, y_train_predict)
    train_f1 = f1_score(y_train, y_train_predict)
    train_precision = precision_score(y_train, y_train_predict)
    train_recall = recall_score(y_train, y_train_predict)
    train_roc_auc = roc_auc_score(y_train, y_train_predict)
    train_classification_report = classification_report(y_train, y_train_predict)

    # Validating the model with testing performance metrics
    test_accuracy = accuracy_score(y_test, y_test_predict)
    test_f1 = f1_score(y_test, y_test_predict)
    test_precision = precision_score(y_test, y_test_predict)
    test_recall = recall_score(y_test, y_test_predict)
    test_roc_auc = roc_auc_score(y_test, y_test_predict)
    test_classification_report = classification_report(y_test, y_test_predict)

    # Printing training performance metrics
    print(ml_model.__class__.__name__)
    print("_" * 50)
    print("Model performance for Training Set")
    print("-" * 50)
    print(f"- Accuracy: {train_accuracy:.4f}")
    print(f"- F1 Score: {train_f1:.4f}")
    print(f"- Precision: {train_precision:.4f}")
    print(f"- Recall: {train_recall:.4f}")
    print(f"- AUC ROC: {train_roc_auc:.4f}")
    print(f"- Classification Report\n{train_classification_report}")
    
    # Printing testing performance metrics
    print("Model performance for Testing Set")
    print("-" * 50)
    print(f"- Accuracy: {test_accuracy:.4f}")
    print(f"- F1 Score: {test_f1:.4f}")
    print(f"- Precision: {test_precision:.4f}")
    print(f"- Recall: {test_recall:.4f}")
    print(f"- AUC ROC: {test_roc_auc:.4f}")
    print(f"- Classification Report\n{test_classification_report}")
    print("-" * 50)
    
    cm_train = confusion_matrix(y_train, y_train_predict)
    cm_test = confusion_matrix(y_test, y_test_predict)

    # Visualize confusion matrix
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))
    
    disp = ConfusionMatrixDisplay(confusion_matrix=cm_train, display_labels=model.classes_)
    disp.plot(ax=axes[0], cmap=plt.cm.Blues)
    axes[0].set_title('Confusion Matrix (Train)')
    axes[0].grid(False)
    
    disp = ConfusionMatrixDisplay(confusion_matrix=cm_test, display_labels=model.classes_)
    disp.plot(ax=axes[1], cmap=plt.cm.Blues)
    axes[1].set_title('Confusion Matrix (Test)')
    axes[1].grid(False)
    
    plt.tight_layout()
    plt.show()
    
    # Compute ROC curve and ROC area (training)
    fpr_train, tpr_train, _ = roc_curve(y_train, y_train_predict)
    fpr_test, tpr_test, _ = roc_curve(y_test, y_test_predict)

    # Plot ROC curve
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))
    
    axes[0].plot(fpr_train, tpr_train, color='darkorange', lw=2, label='ROC curve (AUC = %0.2f)' % train_roc_auc)
    axes[0].plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    axes[0].set_xlim([0.0, 1.0])
    axes[0].set_ylim([0.0, 1.05])
    axes[0].set_xlabel('False Positive Rate')
    axes[0].set_ylabel('True Positive Rate')
    axes[0].set_title('Receiver Operating Characteristic (ROC) Curve (Train)')
    axes[0].legend(loc="lower right")
    
    axes[1].plot(fpr_test, tpr_test, color='darkorange', lw=2, label='ROC curve (AUC = %0.2f)' % test_roc_auc)
    axes[1].plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    axes[1].set_xlim([0.0, 1.0])
    axes[1].set_ylim([0.0, 1.05])
    axes[1].set_xlabel('False Positive Rate')
    axes[1].set_ylabel('True Positive Rate')
    axes[1].set_title('Receiver Operating Characteristic (ROC) Curve (Test)')
    axes[1].legend(loc="lower right")
    
    plt.tight_layout()
    plt.show()
    
    # Append performance metrics
    model_name_l.append(ml_model.__class__.__name__)
    train_accuracy_l.append(round(train_accuracy, 3))
    test_accuracy_l.append(round(test_accuracy, 3))
    train_f1_score_l.append(round(train_f1, 3))
    test_f1_score_l.append(round(test_f1, 3))
    train_precision_l.append(round(train_precision, 3))
    test_precision_l.append(round(test_precision, 3))
    train_recall_l.append(round(train_recall, 3))
    test_recall_l.append(round(test_recall, 3))
    train_auc_roc_l.append(round(train_roc_auc, 3))
    test_auc_roc_l.append(round(test_roc_auc, 3))


def final_results():
    results={"model":model_name_l,
             "train accuracy":train_accuracy_l,"test accuracy":test_accuracy_l,
             "train f1 score":train_f1_score_l,"test f1 score":test_f1_score_l,
            "train precision":train_precision_l, "test precision":test_precision_l,
            "train recall":train_recall_l,"test recall":test_recall_l,
            "train auc roc":train_auc_roc_l,"test auc roc":test_auc_roc_l}

    results_df=pd.DataFrame(results)
    return results_df
```
